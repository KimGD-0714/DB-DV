```python
###### IMPORT #######
import pandas as pd
import json
import FinanceDataReader as fdr
from datetime import datetime
```

```python
###### MAKE MARKET DATA #######

#json data 열기
code_list = []
with open("기업정보.json", encoding='utf-8') as f:
    data = json.load(f)
    for i in range(len(data)):
        code_list.append(str(data[i]["종목코드"]))
        
#print(code_list)

#주식현황 불러오기
df_krx = fdr.StockListing('KOSPI')
year = '2022'

market = pd.DataFrame()

for i in code_list:
    code = i
    new_market = fdr.DataReader(code,year)
    new_market['Code'] = i
    new_market = new_market.reset_index()
    col1 = new_market.columns[-1:].to_list()
    col2 = new_market.columns[:-1].to_list()
    new_col = col1 + col2
    new_market = new_market[new_col]
    market = pd.concat([market,new_market])

market.to_csv(year+'.csv')   
print("Ok")
```

```python
###### MAKE EVENT DATA #######

#json data 열기
code_list = []
ent_list = []
with open("기업정보.json", encoding='utf-8') as f:
    data = json.load(f)
    for i in range(len(data)):
        code_list.append(str(data[i]["종목코드"]))
        ent_list.append(data[i]["기업이름"])
        
event = pd.DataFrame()
event['종목코드'] = code_list
event['종목이름'] = ent_list

event.to_json('event.json', orient = 'records', indent = 1, force_ascii=False)
print('OK')
```

```python
###### MAKE STATUS DATA #######

#json data 열기
code_list = []
major_list = []
have_list = []
ratio_list = []
with open("주주현황.json", encoding='utf-8') as f:
    data = json.load(f)
    for i in range(len(data)):
        code_list.append(str(data[i]["종목코드"]))
        major_list.append(str(data[i]["주요주주"]))
        have_list.append(data[i]["보유주식수"])
        ratio_list.append(data[i]["보유지분"])
        
status = pd.DataFrame()
status['종목코드'] = code_list
status['주요주주'] = major_list
status['보유주식수'] = have_list
status['보유지분'] = ratio_list

status.to_csv('status'+'.csv', index=False, encoding='euc-kr')
print('OK')
```
